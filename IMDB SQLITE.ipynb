```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
import warnings
warnings.filterwarnings('ignore')
```

```python
db="movie.db"
con=sqlite3.connect(db)
cur=con.cursor()
```

```python
cur.execute("select * from movies")
```

```python
movies=cur.fetchall()
```

```python
movies=pd.DataFrame(movies,columns=['id','movie_name','year_of_release','watch_time','movie_rating','metascore','genre','votes','description','directors','Star','gross'])
```
